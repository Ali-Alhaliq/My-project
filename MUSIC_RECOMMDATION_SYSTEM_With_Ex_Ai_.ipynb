```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

def preprocess_text(text):
    # Tokenize the text into words
    tokens = word_tokenize(text)

    # Remove punctuation and lowercase
    tokens = [token.lower() for token in tokens if token not in string.punctuation]

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]



    return tokens

```

```python
df['text'] = [preprocess_text(sentence) for sentence in df['text']]
print(df['text'])
```

```python
df['text'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english')
matrix = tfidvector.fit_transform(df['text'].apply(lambda x: ' '.join(x)))
```

```python
matrix
```

```python
similarity=cosine_similarity(matrix)
```

```python
def recommendation(song_df):
  idx = df[df['song'] == song_df].index[0]
  distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])
  songs = []
  for m_id in distance[1:2]:
    songs.append(df.iloc[m_id[0]].song)
  songi = np.array(songs)
  return songi
```

```python
recommendation('Atom Tan')
```

```python
# Explainable AI function
def explain_recommendation(song_title):
    idx = df[df['song'] == song_title].index[0]
    distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])
    explanations = []

    for m_id in distance[1:2]:
        similar_song_idx = m_id[0]
        similar_song_title = df.iloc[similar_song_idx].song

        # Find the most important words contributing to similarity
        feature_names = tfidvector.get_feature_names_out()
        important_words = np.array(feature_names)[matrix[idx].toarray().flatten().argsort()[-10:]]

        explanations.append({
            "recommended_song": similar_song_title,
            "similarity_score": m_id[1],
            "important_words": important_words
        })

    return explanations

# Get recommendations with explanations
recommendations = recommendation('Atom Tan')
explanations = explain_recommendation('Atom Tan')

print("Recommendations for 'Atom Tan':")
print(recommendations)

print("\nExplanations:")
for explanation in explanations:
    print(f"Song: {explanation['recommended_song']}")
    print(f"Similarity Score: {explanation['similarity_score']}")
    print(f"Important Words: {explanation['important_words']}")
    print("-" * 50)
```
